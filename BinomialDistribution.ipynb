```python
import pandas as pd
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Helpers

def toPercentage(value):
  return round(value * 100, 2)

def plot_binomial_distribution(n, p):
    """
    Plots a binomial distribution

    Parameters:
        n (int): Number of trials.
        p (float): Probability of success.
    """
    # Generate the x values (possible outcomes) and their probabilities
    x_values = np.arange(0, n + 1)
    probabilities = binom.pmf(x_values, n, p)

    data = {'Number of Successes': x_values, 'Probability': probabilities}

    # Create the plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Number of Successes', y='Probability', data=data, color='royalblue')

    # Add labels and title
    plt.title(f"Binomial Distribution (n={n}, p={p})", fontsize=16)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)

size = 200
data = {
  'weight': np.random.randint(low=40, high=251, size=size),
  'age': np.random.randint(low=18, high=76, size=size),
}

df = pd.DataFrame(data)

df
```

Simple Random Sampling

```python
df.sample(n = 5)
```

Cluster Sampling

```python
clusters = df['age'].unique()

# Select random clusters
selected_clusters = np.random.choice(clusters, size=5, replace=False)
selected_clusters.sort()

# Select every row inside the cluster
sample = df[df['age'].isin(selected_clusters)].sort_values(by="age")

print("Selected clusters:", selected_clusters)
print("Sample per cluster:")
sample
```

Stratified Sampling

```python
straction = pd.cut(df['age'], bins=[18, 30, 40, 50, 76], labels=['18-29', '30-39', '40-49', '50+'], right=False)


# 10% of each straction
_, sample = train_test_split(df, test_size=0.1, stratify=straction, random_state=42)

sample.sort_values(by="age")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = np.arange(10, 51)

nsize = 100

# Empty matrix to store the samples
base = np.empty((20, nsize))

# Simple Random Sampling
for i in range(nsize):
  base[:, i] = np.random.choice(population, size=20, replace=True)

# Convert to a pandas DataFrame
base = pd.DataFrame(base)
```

```python
# Calculate the mean of each column
sample_means = base.mean(axis=0).to_numpy()

mean_of_sample_means = np.mean(sample_means)
variance_of_sample_means = np.var(sample_means)

population_mean = np.mean(population)
```

```python
print(f"Mean of the sample means: {mean_of_sample_means}")
print(f"Variance of the sample means: {variance_of_sample_means}")
print(f"Population mean: {population_mean}")

sns.histplot(data=sample_means, bins=12)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Means")
plt.show()
```
